## 3. Feature transformation.
Like standardization of numerical columns and dummy variable or encoding for categorical columns.

```python
# First remove unneccesary columns
df.drop(['Unnamed: 0', 'ID', 'CollegeID', 'CollegeCityID'], axis=1, inplace=True)
df.columns
```

```python
# first: ecode binary categorical columns.

binary_category = df[['Gender', 'CollegeTier', 'CollegeCityTier']]
df.drop(['Gender', 'CollegeTier', 'CollegeCityTier'], axis=1, inplace=True)          # remove these columns from main

pd.set_option('mode.chained_assignment', None)
binary_category.loc[:, 'Gender'] = binary_category['Gender'].map({'m':1, 'f':0})     # map 'm' to 1 and 'f' to 0
binary_category.loc[:, 'CollegeTier'] = binary_category['CollegeTier'] - 1       # also converting college tier to binary

binary_category.head()
```

```python
# Second: encode categorical columns other than binary.
other_category = df[['Designation', 'JobCity', 'Degree', 'Specialization', 'CollegeState', '12board', '10board']]
df.drop(['Designation', 'JobCity', 'Degree', 'Specialization', 'CollegeState', '12board', '10board'], axis=1, inplace=True)

# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)

other_category = pd.DataFrame(encoder.fit_transform(other_category), 
                      columns=encoder.get_feature_names(other_category.columns))

other_category.head()
```

now,
    1. df contains only date columns.
    2. binary_category contains only binary encoded columns.
    3. other_category contains all other categorical column on with OneHotEncoding is applied.
    4. numerical_data contains standardized numerical columns.
